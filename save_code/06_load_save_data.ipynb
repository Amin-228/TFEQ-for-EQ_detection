```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
import os
from utils import data_process, visualize
```

```python
path='/home/Amin/EQ_Place/dataset/init_label/class_8_29/'
True_path, False_path=[], []
o_path='/home/Amin/EQ_Place/dataset/EQ_20210829_130323/'
for f in os.listdir(path+'True/'):
    True_path.append(o_path+f[:-4]+'.csv')
for f in os.listdir(path+'False/'):
    False_path.append(o_path+f[:-4]+'.csv')
```

```python
True_path.sort()
True_path.pop(0)
```

```python
X, Y=[], []
for f in True_path:
    df=pd.read_csv(f)
    df=np.array(df)[:,:3]
    ex,ey,ez=df[:,0],df[:,1],df[:,2]
    ex,ey,ez=ex-np.mean(ex), ey-np.mean(ey), ez-np.mean(ez)
    X.append([ex,ey,ez])
#     X.append([np.array([ex]),np.array([ey]),np.array([ez])])
    Y.append(1)
for f in False_path:
    df=pd.read_csv(f)
    df=np.array(df)[:,:3]
    ex,ey,ez=df[:,0],df[:,1],df[:,2]
    ex,ey,ez=ex-np.mean(ex), ey-np.mean(ey), ez-np.mean(ez)
    X.append([ex,ey,ez])
    Y.append(0)
X=np.array(X)
# X=X.reshape(X.shape[0],-1)
Y=np.array(Y)
```

```python
visualize.draw_picture(X[0], show=True, save_path=None, lim=0.01,save=False)
```
